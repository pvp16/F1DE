```redshift
select driverName, count(*) as races_won
from factresults f
inner join dimdrivers d on d.driverkey = f.driverkey
where position = 1
group by driverName
order by races_won desc

select driverName, count(*) as Poles
from factresults f
inner join dimdrivers d on d.driverkey = f.driverkey
where grid = 1
group by driverName
order by Poles desc

select latitude,longitude,country,locality, count(*) as races_held
from (select distinct r.racekey,latitude,longitude,datekey,country,locality
from dimraces d 
inner join factresults r on r.racekey = d.racekey) t
group by latitude, longitude,country,locality
order by races_held desc



select constructorname,year, count(case when position in (1,2,3) then 1 end) as podiums
from factresults f
inner join dimconstructors c on c.constructorkey = f.constructorkey
inner join dimdate d on d.datekey = f.datekey
where constructorid in ('ferrari','red_bull', 'mercedes','mclaren') and year between 2000 and 2022
group by year,constructorname
order by year

select drivername, round,sum(points) over(partition by drivername order by round rows between unbounded preceding and current row) as cpoints
from factresults f
inner join dimdrivers d on d.driverkey = f.driverkey
where drivername in ('Lewis Hamilton','Max Verstappen') and datekey like '%2021%'

select sum(points), drivername
from factresults f
inner join dimdrivers d on d.driverkey = f.driverkey
where datekey like '%2021%'
group by drivername

select drivernationality, count(*) as total
from (select distinct drivernationality, drivername
from dimdrivers) t
group by drivernationality
order by total desc

select constructornationality, count(*) as total
from (select distinct constructornationality, constructorname
from dimconstructors) t
group by constructornationality
order by total desc

select distinct constructornationality, constructorid
from dimconstructors
order by constructornationality, constructorid
```
